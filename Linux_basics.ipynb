```python
# CREATE A FASTA FILE using Linux
!echo ">seq1 Test sequence" > bioinformatics/data/test.fasta
!echo "ATGCTTACGGATCAAGCTA" >> bioinformatics/data/test.fasta
!echo ">seq2 Second sequence" >> bioinformatics/data/test.fasta
!echo "GCTAGCTAGCTAGCTAGCT" >> bioinformatics/data/test.fasta

# Display it
!cat bioinformatics/data/test.fasta
```

```python
# ============================================
# ESSENTIAL LINUX COMMANDS — ALL IN ONE CELL
# ============================================

# 1. COUNT LINES IN A FILE
# wc = word count
# wc -l = count lines only
!wc -l bioinformatics/data/test.fasta

# 2. SEARCH INSIDE A FILE
# grep = search for pattern
# grep ">" searches for header lines in FASTA
!echo "--- Searching for headers ---"
!grep ">" bioinformatics/data/test.fasta

# 3. COPY A FILE
# cp = copy
!cp bioinformatics/data/test.fasta bioinformatics/data/test_backup.fasta
!echo "--- Files in data folder ---"
!ls bioinformatics/data/

# 4. CHECK FILE SIZE
# du = disk usage
# -h = human readable (shows KB, MB not bytes)
!echo "--- File sizes ---"
!du -h bioinformatics/data/test.fasta

# 5. FIRST FEW LINES OF FILE
# head = show first lines
!echo "--- First 2 lines ---"
!head -2 bioinformatics/data/test.fasta

# 6. LAST FEW LINES OF FILE
# tail = show last lines
!echo "--- Last 2 lines ---"
!tail -2 bioinformatics/data/test.fasta
```

```python
# Create a bigger FASTA file first
!echo ">gene1 BRCA1" > bioinformatics/data/genes.fasta
!echo "ATGCTTACGGATCAAGCTA" >> bioinformatics/data/genes.fasta
!echo ">gene2 TP53" >> bioinformatics/data/genes.fasta
!echo "GCTAGCTAGCTAGCTAGCT" >> bioinformatics/data/genes.fasta
!echo ">gene3 INS" >> bioinformatics/data/genes.fasta
!echo "ATGGCCCTGTGGATGCGCC" >> bioinformatics/data/genes.fasta
!echo ">gene4 EGFR" >> bioinformatics/data/genes.fasta
!echo "ATGCGACCCTCCGGGACGG" >> bioinformatics/data/genes.fasta

# COUNT SEQUENCES IN FASTA FILE
# grep ">" finds all headers
# wc -l counts them
!echo "Number of sequences in file:"
!grep ">" bioinformatics/data/genes.fasta | wc -l

# SEARCH AND SHOW ONLY MATCHING LINES
!echo "Headers containing gene name:"
!grep ">" bioinformatics/data/genes.fasta
```

```python
# WRITE A BASH SCRIPT
# This script analyzes a FASTA file automatically

bash_script = """#!/bin/bash
# My first bash script
# Analyzes a FASTA file

echo "=============================="
echo "  FASTA FILE ANALYSIS REPORT"
echo "=============================="

FILE=$1   # $1 means first argument you pass to script

echo "File analyzed: $FILE"
echo ""

# Count sequences
SEQ_COUNT=$(grep ">" $FILE | wc -l)
echo "Total sequences: $SEQ_COUNT"

# Show all sequence names
echo ""
echo "Sequence names:"
grep ">" $FILE

# Count total lines
LINES=$(wc -l < $FILE)
echo ""
echo "Total lines in file: $LINES"

echo "=============================="
echo "Analysis complete!"
echo "=============================="
"""

# Save the script to a file
with open("bioinformatics/scripts/analyze_fasta.sh", "w") as f:
    f.write(bash_script)

print("✅ Bash script created!")

# Make it executable
!chmod +x bioinformatics/scripts/analyze_fasta.sh

# RUN the script on our genes file
!bash bioinformatics/scripts/analyze_fasta.sh bioinformatics/data/genes.fasta
```

```python
# COMBINING LINUX AND PYTHON IN ONE WORKFLOW

import subprocess

# Run Linux command from Python
# capture_output=True saves the result
result = subprocess.run(
    ["grep", ">", "bioinformatics/data/genes.fasta"],
    capture_output = True,
    text           = True
)

# Get the output as a Python string
headers = result.stdout.strip().split("\n")

print("Found these sequences:")
for i, header in enumerate(headers):
    print(f"  Sequence {i+1}: {header}")

print(f"\nTotal: {len(headers)} sequences found")
```
