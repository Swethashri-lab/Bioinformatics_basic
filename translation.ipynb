```python
# ============================================
# COMPLETE DNA PIPELINE — Days 1 through 6
# ============================================
!pip install Biopython
import Bio
from Bio.Seq import Seq
from Bio.SeqUtils import gc_fraction

def complete_pipeline(dna, sequence_name="Unknown"):

    dna = dna.upper().replace(" ", "").replace("\n", "")
    seq = Seq(dna)

    print("=" * 60)
    print(f"  COMPLETE ANALYSIS: {sequence_name}")
    print("=" * 60)

    # Day 1 — nucleotide counts
    print("\n📊 NUCLEOTIDE COMPOSITION")
    print(f"  Sequence : {dna[:50]}{'...' if len(dna)>50 else ''}")
    print(f"  Length   : {len(dna)} bases")
    print(f"  A: {dna.count('A')}  T: {dna.count('T')}  G: {dna.count('G')}  C: {dna.count('C')}")

    # Day 2 — GC content
    gc = round(gc_fraction(seq) * 100, 2)
    print(f"\n🔬 GC CONTENT: {gc}%")
    if gc > 60:
        print("  Stability: 🔴 High GC — very stable")
    elif gc < 40:
        print("  Stability: 🔵 Low GC — less stable")
    else:
        print("  Stability: 🟢 Normal range")

    # Day 3 — transcription and reverse complement
    rna      = seq.transcribe()
    rev_comp = seq.reverse_complement()
    print(f"\n🧬 TRANSCRIPTION")
    print(f"  RNA      : {str(rna)[:50]}...")
    print(f"  Rev Comp : {str(rev_comp)[:50]}...")

    # Day 6 — translation
    protein = seq.translate(to_stop=True)
    print(f"\n🔑 TRANSLATION")
    print(f"  Protein  : {protein}")
    print(f"  Length   : {len(protein)} amino acids")

    print("=" * 60)


# Test with real sequences
complete_pipeline(
    "ATGAAAGCAATTTTCGTACTGAAAGGTTTTGTTTAA",
    "Test Gene 1"
)

print()

complete_pipeline(
    "ATGGCCCTGTGGATGCGCCTCCTGCCCCTGCTGGCGCTGCTGGCCCTCTGGTAA",
    "Human Insulin Fragment"
)
```
