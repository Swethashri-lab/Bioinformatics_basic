```python
# Import Biopython tools
from Bio.Seq import Seq

# Create a sequence object
# This is NOT just a string anymore
# It's a biological sequence with superpowers

my_seq = Seq("ATGCTTACGGATC")

print("=" * 45)
print("BIOPYTHON SEQUENCE ANALYSIS")
print("=" * 45)

# Length
print("Sequence    :", my_seq)
print("Length      :", len(my_seq), "bases")

# Nucleotide counts
print("A count     :", my_seq.count("A"))
print("T count     :", my_seq.count("T"))
print("G count     :", my_seq.count("G"))
print("C count     :", my_seq.count("C"))

# GC content — ONE LINE instead of your Day 2 formula
from Bio.SeqUtils import gc_fraction
gc = round(gc_fraction(my_seq) * 100, 2)
print("GC Content  :", gc, "%")

# Transcription — ONE LINE instead of .replace("T","U")
rna = my_seq.transcribe()
print("RNA         :", rna)

# Reverse complement — ONE LINE instead of your Day 3 loop
rev_comp = my_seq.reverse_complement()
print("Rev Comp    :", rev_comp)

# Translation — DNA to Protein (NEW! You haven't done this yet)
protein = my_seq.translate()
print("Protein     :", protein)

print("=" * 45)
```

```python
# First recreate the sample.fasta file
# (Colab forgets files between sessions)

fasta_content = """>seq1 Human BRCA1 fragment
ATGGATTTATCTGCTCTTCGCGTTGAAGAAGTACAAAATGTCATTAATGCTATGCAGA
AAATCTTAGAGTGTCCCATCTGTCTGGAGTTGATCAAGGAACCTGTCTCCACAAAGTGT
>seq2 Human TP53 tumor suppressor
ATGGAGGAGCCGCAGTCAGATCCTAGCGTTGAATCTGAGGAATGCTTCGTCCTTTCACT
CCATCGATGCTGTTCCGGGAGCAGATAAGCGGGCCGGTCTTCCCTTCCCAGAAACCCCT
>seq3 Mouse beta-actin
ATGGATGATGATATCGCCGCGCTCGTCGTCGACAACGGCTCCGGCATGTGCAAAGCCGG
CTTCGCGGGCGACGATGCCCCGAGGGCCGTCTTCCCCTCCATCGTGGGGCGCCCCAGGC"""

with open("sample.fasta", "w") as file:
    file.write(fasta_content)

print("✅ sample.fasta ready!")
```

```python
# Read FASTA file using Biopython
# Compare this to your 40-line Day 4 parser!

from Bio import SeqIO

print("=" * 55)
print("READING FASTA WITH BIOPYTHON")
print("=" * 55)

# SeqIO.parse reads the file and gives you records one by one
for record in SeqIO.parse("sample.fasta", "fasta"):

    # Each record has three main things:
    # record.id       = sequence ID
    # record.description = full header line
    # record.seq     = the actual sequence

    from Bio.SeqUtils import gc_fraction

    gc = round(gc_fraction(record.seq) * 100, 2)

    print()
    print("🧬 ID          :", record.id)
    print("   Description :", record.description)
    print("   Length      :", len(record.seq), "bases")
    print("   GC Content  :", gc, "%")
    print("   Sequence    :", str(record.seq)[:40], "...")
    print("   RNA         :", str(record.seq.transcribe())[:40], "...")
    print("-" * 55)

print()
print("✅ Done! All sequences analyzed.")
```

```python
# Download real gene from NCBI database
# We're downloading the Human Insulin gene

from Bio import Entrez, SeqIO

# Always tell NCBI who you are — required by their rules
Entrez.email = "sweth102003@gmail.com"    # replace with your email

print("⏳ Connecting to NCBI database...")
print("Downloading Human Insulin gene...")

# Fetch the sequence
# NM_000207 is the official ID for human insulin gene
handle = Entrez.efetch(
    db       = "nucleotide",    # which database to search
    id       = "NM_000207",     # the gene ID
    rettype  = "fasta",         # get it in FASTA format
    retmode  = "text"           # as plain text
)

# Parse what we downloaded
record = SeqIO.read(handle, "fasta")
handle.close()

print()
print("=" * 55)
print("HUMAN INSULIN GENE — DOWNLOADED FROM NCBI")
print("=" * 55)
print("ID          :", record.id)
print("Description :", record.description[:60], "...")
print("Length      :", len(record.seq), "bases")

from Bio.SeqUtils import gc_fraction
gc = round(gc_fraction(record.seq) * 100, 2)
print("GC Content  :", gc, "%")
print()
print("First 100 bases of real human insulin:")
print(str(record.seq)[:100])
print("=" * 55)
print()
print("✅ Real gene downloaded successfully from NCBI!")
```

```python
# Save the real gene to a FASTA file on Colab

with open("human_insulin.fasta", "w") as output:
    SeqIO.write(record, output, "fasta")

print("✅ Human insulin gene saved as human_insulin.fasta")

# Verify by reading it back
print()
print("Verification — reading saved file:")
for rec in SeqIO.parse("human_insulin.fasta", "fasta"):
    print("ID     :", rec.id)
    print("Length :", len(rec.seq), "bases")
```
