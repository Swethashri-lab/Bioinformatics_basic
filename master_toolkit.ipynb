```python
# ============================================
# BIOINFORMATICS TOOLKIT
# Complete DNA Analysis Pipeline
# Built by: Your Name
# Days 1-6 Combined
# ============================================

!pip install biopython -q   # -q means quiet — less output printed

from Bio.Seq import Seq
from Bio.SeqUtils import gc_fraction

# ============================================
# CODON TABLE
# ============================================

codon_table = {
    "UUU":"Phe", "UUC":"Phe", "UUA":"Leu", "UUG":"Leu",
    "CUU":"Leu", "CUC":"Leu", "CUA":"Leu", "CUG":"Leu",
    "AUU":"Ile", "AUC":"Ile", "AUA":"Ile", "AUG":"Met",
    "GUU":"Val", "GUC":"Val", "GUA":"Val", "GUG":"Val",
    "UCU":"Ser", "UCC":"Ser", "UCA":"Ser", "UCG":"Ser",
    "CCU":"Pro", "CCC":"Pro", "CCA":"Pro", "CCG":"Pro",
    "ACU":"Thr", "ACC":"Thr", "ACA":"Thr", "ACG":"Thr",
    "GCU":"Ala", "GCC":"Ala", "GCA":"Ala", "GCG":"Ala",
    "UAU":"Tyr", "UAC":"Tyr", "CAU":"His", "CAC":"His",
    "CAA":"Gln", "CAG":"Gln", "AAU":"Asn", "AAC":"Asn",
    "AAA":"Lys", "AAG":"Lys", "GAU":"Asp", "GAC":"Asp",
    "GAA":"Glu", "GAG":"Glu", "UGU":"Cys", "UGC":"Cys",
    "UGG":"Trp", "CGU":"Arg", "CGC":"Arg", "CGA":"Arg",
    "CGG":"Arg", "AGA":"Arg", "AGG":"Arg", "AGU":"Ser",
    "AGC":"Ser", "GGU":"Gly", "GGC":"Gly", "GGA":"Gly",
    "GGG":"Gly", "UAA":"STOP","UAG":"STOP","UGA":"STOP"
}


# ============================================
# TOOL 1 — NUCLEOTIDE COUNTER (Day 1)
# ============================================

def count_nucleotides(dna):
    dna = dna.upper().replace(" ", "")
    print("[ NUCLEOTIDE COMPOSITION ]")
    print(f"  A: {dna.count('A')}")
    print(f"  T: {dna.count('T')}")
    print(f"  G: {dna.count('G')}")
    print(f"  C: {dna.count('C')}")
    print(f"  Total Length: {len(dna)} bases")


# ============================================
# TOOL 2 — GC CONTENT (Day 2)
# ============================================

def get_gc_content(dna):
    dna = dna.upper().replace(" ", "")
    gc  = round((dna.count("G") + dna.count("C")) / len(dna) * 100, 2)
    print("[ GC CONTENT ]")
    print(f"  GC Content: {gc}%")
    if gc > 60:
        print("  Stability : High GC - very stable DNA 🔴")
    elif gc < 40:
        print("  Stability : Low GC - less stable DNA 🔵")
    else:
        print("  Stability : Normal GC range 🟢")
    return gc


# ============================================
# TOOL 3 — TRANSCRIPTION + REVERSE COMPLEMENT (Day 3)
# ============================================

def transcribe_and_complement(dna):
    dna      = dna.upper().replace(" ", "")
    seq      = Seq(dna)
    rna      = seq.transcribe()
    rev_comp = seq.reverse_complement()
    print("[ TRANSCRIPTION + COMPLEMENT ]")
    print(f"  RNA              : {str(rna)[:50]}...")
    print(f"  Reverse Complement: {str(rev_comp)[:50]}...")


# ============================================
# TOOL 4 — FASTA PARSER (Day 4)
# ============================================

def parse_fasta(filename):
    sequences    = {}
    current_name = ""
    current_seq  = ""
    with open(filename, "r") as file:
        for line in file:
            line = line.strip()
            if line == "":
                continue
            if line.startswith(">"):
                if current_name != "":
                    sequences[current_name] = current_seq
                current_name = line[1:]
                current_seq  = ""
            else:
                current_seq = current_seq + line
        if current_name != "":
            sequences[current_name] = current_seq
    return sequences


# ============================================
# TOOL 5 — PROTEIN TRANSLATION (Day 6)
# ============================================

def translate_dna(dna):
    dna     = dna.upper().replace(" ", "")
    seq     = Seq(dna)
    protein = seq.translate(to_stop=True)
    print("[ PROTEIN TRANSLATION ]")
    print(f"  Protein : {protein}")
    print(f"  Length  : {len(protein)} amino acids")


# ============================================
# MASTER PIPELINE — RUNS ALL TOOLS TOGETHER
# ============================================

def run_complete_pipeline(dna, sequence_name="My Sequence"):

    dna = dna.upper().replace(" ", "").replace("\n", "")

    print()
    print("=" * 60)
    print(f"  BIOINFORMATICS TOOLKIT — FULL ANALYSIS")
    print(f"  Sequence: {sequence_name}")
    print("=" * 60)

    print()
    count_nucleotides(dna)

    print()
    get_gc_content(dna)

    print()
    transcribe_and_complement(dna)

    print()
    translate_dna(dna)

    print()
    print("=" * 60)
    print("  ANALYSIS COMPLETE")
    print("=" * 60)
    print()


# ============================================
# RUN THE PIPELINE
# ============================================

run_complete_pipeline(
    "ATGAAAGCAATTTTCGTACTGAAAGGTTTTGTTTAA",
    "Test Sequence 1"
)

run_complete_pipeline(
    "ATGGCCCTGTGGATGCGCCTCCTGCCCCTGCTGGCGCTGCTGGCCCTCTGGTAA",
    "Human Insulin Fragment"
)
```
