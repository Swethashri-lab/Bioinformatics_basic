```python
# SEE ALL CONDA COMMANDS AND WHAT THEY DO

print("""
CONDA COMMANDS REFERENCE
========================

Creating:
  conda create -n myenv python=3.9
  → Creates new environment called myenv with Python 3.9

Activating:
  conda activate myenv
  → Enter/switch to an environment

Deactivating:
  conda deactivate
  → Leave current environment, go back to base

Installing packages:
  conda install -n myenv biopython
  → Install biopython into myenv

Listing environments:
  conda env list
  → Show all environments you have

Listing packages:
  conda list -n myenv
  → Show all packages in myenv

Deleting environment:
  conda env remove -n myenv
  → Delete entire environment

Exporting environment:
  conda env export -n myenv > environment.yml
  → Save environment to a file (share with others)

Creating from file:
  conda env create -f environment.yml
  → Recreate environment from saved file
""")
```

```python

# INSTALL PACKAGES INTO OUR ENVIRONMENT
# Three ways to install into a specific environment:

# Method 1 — conda install (preferred for bioinformatics tools)
!conda install -n bioinfo_env biopython -y

# Method 2 — pip install into specific environment
!conda run -n bioinfo_env pip install pandas numpy matplotlib

# Verify what's installed
!conda list -n bioinfo_env
```

```python
# LIST ALL ENVIRONMENTS
!conda env list

# Output will show:
# base          *  /usr/local
# bioinfo_env      /usr/local/envs/bioinfo_env
```

```python
# CREATE environment.yml manually
# This is what you'll put on GitHub with every project

env_content = """name: bioinfo_env
channels:
  - conda-forge
  - bioconda
  - defaults
dependencies:
  - python=3.9
  - biopython=1.83
  - numpy=1.24
  - pandas=2.0
  - matplotlib=3.7
  - pip
  - pip:
    - condacolab
"""

with open("environment.yml", "w") as f:
    f.write(env_content)

print("✅ environment.yml created!")
!cat environment.yml
```

```python
# WHAT ARE CHANNELS?

print("""
Channels = where conda looks for packages

conda-forge:
  → Community maintained
  → Largest collection of packages
  → Most up to date versions
  → Use this first

bioconda:
  → Specifically for BIOINFORMATICS tools
  → Contains: samtools, hisat2, fastqc, star
  → Essential for NGS pipelines
  → Use this for all bio tools

defaults:
  → Anaconda's official channel
  → Smaller collection
  → Most stable but sometimes outdated

Priority order in our yml:
conda-forge → bioconda → defaults
Conda searches in this order
""")
```

```python
# CREATE ENVIRONMENT FROM YML FILE
# This is how you'll set up every new project

!conda env create -f environment.yml

# Verify it worked
!conda env list
```

```python
# REAL WORLD CONDA WORKFLOW
# This is what professional bioinformaticians do

print("""
STARTING A NEW PROJECT:
───────────────────────
Step 1: Create environment
  conda create -n project_name python=3.9

Step 2: Activate it
  conda activate project_name

Step 3: Install what you need
  conda install -n project_name biopython pandas

Step 4: Work on your project
  python my_analysis.py

Step 5: Export environment when done
  conda env export -n project_name > environment.yml

Step 6: Push environment.yml to GitHub
  git add environment.yml
  git commit -m "Add conda environment"
  git push


WHEN SOMEONE ELSE WANTS TO RUN YOUR CODE:
──────────────────────────────────────────
Step 1: Clone your GitHub repository
  git clone your_repo_url

Step 2: Create environment from your yml
  conda env create -f environment.yml

Step 3: Activate it
  conda activate project_name

Step 4: Run your code
  python my_analysis.py

Everything works perfectly — same versions, same packages.
""")
```

```python
# NGS PIPELINE ENVIRONMENT
# We'll use this starting Week 3

ngs_env = """name: ngs_pipeline
channels:
  - conda-forge
  - bioconda
  - defaults
dependencies:
  - python=3.9
  - biopython=1.83
  - numpy=1.24
  - pandas=2.0
  - matplotlib=3.7
  - fastqc
  - trimmomatic
  - hisat2
  - samtools
  - subread
"""

with open("ngs_environment.yml", "w") as f:
    f.write(ngs_env)

print("✅ NGS environment file created!")
print()
print("This environment contains:")
print("  - Python tools: biopython, pandas, numpy")
print("  - QC tools    : fastqc, trimmomatic")
print("  - Alignment   : hisat2")
print("  - BAM tools   : samtools")
print("  - Counting    : subread (featureCounts)")
print()
print("We will USE this environment in Week 3!")
```
